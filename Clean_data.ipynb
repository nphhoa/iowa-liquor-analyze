```python
'''
Store Location column has many unique data values due to differences from rounding decimal places in the store's location.
-> Consider the geographical location as a type of data with relatively accurate precision, as there are many issues that can arise with this data, such as rounding errors, some places having the same longitude but different latitudes and vice versa, etc. Therefore, it is not possible to demand absolute accuracy in this information, but rather it will need to be retained with a certain level of relative accuracy, while also considering this during the data utilization process.
-> Use the data from 'Store Location' to extract two new variables: 'Longitude' and 'Latitude', then remove the 'Store Location' column from the dataset.
'''
```

```python
'''
It can be observed that the categorical variables contain a large number of unique values. This approach will not be objective if we use visual representations to depict the information of these variables.
Instead, I will combine the information provided by these categorical variables in the diagnostic analysis section.
'''
```

```python
# Numerical features
num = ['State Bottle Cost', 'State Bottle Retail', 'Bottles Sold', 'Sale (Dollars)', 'Volume Sold (Liters)', 'Volume Sold (Gallons)']
```

```python
# Function to visualize and analyze a univariate continuous variable
def univariate_visualization_analysis_continuous_variable(df, column_name):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # KDE plot
    sns.kdeplot(df[column_name], ax=axes[0])
    axes[0].set_title(f'{column_name}')
    
    # Box plot
    sns.boxplot(x=df[column_name], ax=axes[1])
    axes[1].set_title(f'{column_name}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Function for univariate analysis on continuous variables
def univariate_analysis_continuous_variable(df, column_name):
    desc = df[column_name].describe()
    print('Describe:')
    print(desc)
    print('Name:', df[column_name].name, ', dtype:', df[column_name].dtype)
    print('Mode:', df[column_name].mode()[0])
    print('Nunique:', df[column_name].nunique())
    print('Range:', desc['max'] - desc['min'])
    IQR = desc['75%'] - desc['25%']
    print('IQR:', IQR)
    print('Var:', df[column_name].var())
    print('Skew:', df[column_name].skew())
    print('Kurtosis:', df[column_name].kurtosis())
```

```python
def check_outlier(df, column_name):
    q1 = df[column_name].quantile(0.25)
    q3 = df[column_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    n_O_upper = sum(df[column_name] > upper_bound)
    n_L_upper = sum(df[column_name] < lower_bound)
    outliers_per = (n_O_upper + n_L_upper) / len(df) * 100
    
    return n_O_upper, n_L_upper, outliers_per
```

```python
num = ['State Bottle Cost', 'State Bottle Retail', 'Bottles Sold', 'Sale (Dollars)', 'Volume Sold (Liters)']
for col in num:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(df, col)
    univariate_analysis_continuous_variable(df, col)
    n_O_upper, n_L_upper, outliers_per = check_outlier(df, col)
    print(f'Number of upper outliers: {n_O_upper}')
    print(f'Number of lower outliers: {n_L_upper}')
    print(f'Percentage of outliers: {outliers_per:.1f} %')
    print('\n')
    print('*' * 60 ,'\n')
```

```python
'''
1. Analysis of Dispersion and Outliers:
Statistical measures:
- High standard deviation (std) and variance (var) indicate a wide dispersion of data.
- High skewness (skew) and kurtosis (kurtosis) indicate a non-symmetrical distribution and the presence of outliers.
- Large range (range) indicates a significant gap between the maximum (max) and minimum (min) values.
Visualizations:
- Distribution (distplot) and boxplot (boxplot) clearly show the large gap between the maximum and minimum data points.
- The majority of the data is concentrated at very low values, while a small number of outliers (around 4.1-6.5%) have significantly higher values than the rest.
2. Justification for Keeping Outliers:
Analysis objective: The project focuses on customer segmentation (Customer Segmentation). Therefore, outliers are important in explaining the characteristics of each customer group (e.g., stores with high order volumes, different alcohol consumption patterns across regions).
3. Handling Outliers in Descriptive Analysis:
Temporarily remove outliers to observe the distribution of the majority of the data.
The decision to remove or keep outliers will be carefully considered based on the analysis objective and the significance of each variable.
'''
```

```python
# Create new dataframe to drop outliers
data_no_outliers = df[num].copy()

# List percentile of data no outliers for each feature
list_per = []

# Create a loop to drop outliers
for var in num:
    # Calculate Q1 & Q3
    Q1 = np.percentile(data_no_outliers[var], 25)
    Q3 = np.percentile(data_no_outliers[var], 75)
    
    # Calculate IQR
    IQR = Q3 - Q1

    # Find Lower Whisker & Upper Whisker
    LW = Q1 - 1.5 * IQR
    UW = Q3 + 1.5 * IQR

    # Get outliers value
    outliers = data_no_outliers[(data_no_outliers[var] < LW) | (data_no_outliers[var] > UW)]

    # Calculate percent of data when drop outliers for each feature
    per = round((1 - len(outliers) / len(data_no_outliers[var])) * 100, 0)
    
    # Add that percent to list_per
    list_per.append(per)
    
    # Drop outliers
    data_no_outliers = data_no_outliers.drop(index=outliers.index)
    data_no_outliers.reset_index(drop=True, inplace=True)
```

```python
i = 0
for col in num:
    print('Feature:', col)
    univariate_visualization_analysis_continuous_variable(data_no_outliers, col)
    univariate_analysis_continuous_variable(data_no_outliers, col)
    n_O_upper, n_L_upper, outliers_per = check_outlier(data_no_outliers, col)
    print("Percentage data of",col,"when drop outliers:", list_per[i],"%")
    print('\n')
    print('=' * 120 ,'\n')
    i += 1
```
